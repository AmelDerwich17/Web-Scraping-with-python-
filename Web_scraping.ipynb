```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://forecast.weather.gov/MapClick.php?lat=34.05349000000007&lon=-118.24531999999999#.XgC1lEdKg2w')
soup = BeautifulSoup(page.content,'html.parser')
print(soup.find_all('img'))
```

```python
week = soup.find(id='seven-day-forecast-body')
```

```python
print(week)
```

```python
print(week.find_all(class_='forecast-tombstone'))
```

```python
items = week.find_all(class_='tombstone-container')
```

```python
print(items[1]) #Get the second item with class = "tombstone-container"
```

```python
print(items[1].find(class_='period-name').get_text())
```

```python
print(items[1].find(class_='short-desc').get_text())
```

```python
print(items[1].find(class_='temp').get_text())
```

```python
#we will write a loop to get all period name in a list
period_names = [item.find(class_='period-name').get_text() for item in items]
print(period_names)
```

```python
#we will repeat the same thing to short-desc and temp
short_desc = [item.find(class_='short-desc').get_text() for item in items]
temp = [item.find(class_='temp').get_text() for item in items]
print(short_desc)
print(temp)
```

```python
weather_stuff = pd.DataFrame(
    {
    'period': period_names,
    'short_descriptions': short_desc,
    'temperature': temp,
    })
print(weather_stuff)
```

```python
#export to csv file 
weather_stuff.to_csv('Weather.csv')
```
